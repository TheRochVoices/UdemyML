```python
import pandas as pd
import numpy as np
from apyori import apriori
```

```python
df = pd.read_csv('Market_Basket_Optimisation.csv', header=None)
```

```python
df.head()
```

```python
#apriori accepts list of lists. Converting Dataframe into list
tr = df.values.tolist()
```

```python
for i in range(0, df.shape[0]):
    for j in range(0, df.shape[1]):
        tr[i][j] = str(tr[i][j])
```

```python
rules = apriori(tr, min_support=.003, min_confidence=.2, min_lift=3, min_length=3)
```

```python
res = list(rules)
```
