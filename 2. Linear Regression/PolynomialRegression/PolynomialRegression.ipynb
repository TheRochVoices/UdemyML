```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataSet = pd.read_csv('Position_Salaries.csv')
```

```python
lvl = dataSet.iloc[:, 1:2].values
slry = dataSet.iloc[:, 2].values
```

```python
plt.scatter(lvl, slry, color='red')
plt.title('Level vs Salary')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```

# Trying Polynomial Regression with degree 2

```python
from sklearn.preprocessing import PolynomialFeatures

from sklearn.linear_model import LinearRegression
```

```python
poly = PolynomialFeatures(degree=2)
lvl_poly = poly.fit_transform(lvl)


reg2 = LinearRegression()
reg2.fit(lvl_poly, slry)
```

```python
plt.scatter(lvl, slry, color='red')
plt.plot(lvl, reg2.predict(lvl_poly))
plt.title('Level vs Salary')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```

# Increasing degree to 2 to achieve better results

```python
poly = PolynomialFeatures(degree=3)
lvl_poly = poly.fit_transform(lvl)


reg3 = LinearRegression()
reg3.fit(lvl_poly, slry)
```

```python
plt.scatter(lvl, slry, color='red')
plt.plot(lvl, reg3.predict(lvl_poly))
plt.title('Level vs Salary')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```

# Better than degree=3. Trying with 4

```python
poly = PolynomialFeatures(degree=4)
lvl_poly = poly.fit_transform(lvl)


reg4 = LinearRegression()
reg4.fit(lvl_poly, slry)
```

```python
plt.scatter(lvl, slry, color='red')
plt.plot(lvl, reg4.predict(lvl_poly))
plt.title('Level vs Salary')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```

```python
slry_pred = reg4.predict(lvl_poly)
print('Salary \t\t Predicted')
for i, j in zip(slry, slry_pred):
    print(i, '\t\t', j)
```
