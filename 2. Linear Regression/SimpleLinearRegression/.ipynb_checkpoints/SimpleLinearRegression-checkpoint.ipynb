```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataSet = pd.read_csv('Salary_Data.csv')
years = dataSet.iloc[:, :-1].values
salary = dataSet.iloc[:, 1].values
```

```python
from sklearn.cross_validation import train_test_split
```

```python
y_train, y_test, s_train, s_test = train_test_split(years, salary, test_size = 0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
regression.fit(y_train, s_train)
```

```python
plt.scatter(y_train, s_train, color='red')
plt.plot(y_train, regression.predict(y_train))
plt.title('Linear Regression (Train Data)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
plt.scatter(y_test, s_test, color='red')
plt.plot(y_train, regression.predict(y_train))
plt.title('Linear Regression (Test Data)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
